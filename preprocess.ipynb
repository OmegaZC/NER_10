```python
# 多列dataframe
# 字典初始化
data_list = []
with open(input_path , 'r' , encoding='utf-8') as f:
    # 读取每一行
    for line in f:
        # 如果这一行是换行符就跳过
        if line.count('\n') == len(line):
            continue
        
        
        data_list.append([line.strip('\n').strip()[5:].strip() if line.strip('\n').strip().find('\t') != -1 else line.strip('\n').strip()])
```

```python


df = pd.DataFrame(data_list , columns=['yaoqiu'])
```

```python
df.head()
df.describe(include='all')
```

```python
df.drop_duplicates(inplace=True)
df.describe()
df.head()
```

```python
# (x.encode() - len(x)) // 2：计算中文个数 编码后是三个字符长度
df['CN_count'] = df['yaoqiu'].apply(lambda x : (len(x.encode()) - len(x)) // 2)
# 也可以正则识别法
# import re
# re.findall('([\u4e00-\u9fa5])' , x)
# # 返回的是中文每个字为一个元素的列表
```
